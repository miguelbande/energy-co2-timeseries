```python
import pandas as pd

ruta_archivo = 'COMPLETAR CON LA RUTA DE LOS ARCHIVOS'
```

```python
# Vamos a importar todos los datos de los distintos archivos de excel

df_2015_2017 = pd.read_excel(ruta_archivo + 'MHLV_data-2015-2019.xlsx', sheet_name='2015-2017')
df_2018_2019 = pd.read_excel(ruta_archivo + 'MHLV_data-2015-2019.xlsx', sheet_name='2018-2019')
df_2019 = pd.read_excel(ruta_archivo + 'monthly_hourly_load_values_2019.xlsx', sheet_name='Sheet1')
df_2020 = pd.read_excel(ruta_archivo + 'monthly_hourly_load_values_2020.xlsx', sheet_name='Sheet1')
df_2021 = pd.read_excel(ruta_archivo + 'monthly_hourly_load_values_2021.xlsx', sheet_name='Sheet1')
df_2022 = pd.read_excel(ruta_archivo + 'monthly_hourly_load_values_2022.xlsx', sheet_name='Sheet1')
df_2023 = pd.read_excel(ruta_archivo + 'monthly_hourly_load_values_2023.xlsx', sheet_name='Sheet1')
df_2024 = pd.read_excel(ruta_archivo + 'monthly_hourly_load_values_2024.xlsx', sheet_name='Sheet1')
df_2025 = pd.read_excel(ruta_archivo + 'monthly_hourly_load_values_2025.xlsx', sheet_name='Sheet1')

```

```python
# Como ocupan mucho los dataframes, vamos a quedarnos solamente con las columnas que nos interesan: 'DateShort', 'Value' y 'CountryCode'

df_2015_2017 = df_2015_2017[['DateShort', 'Value', 'CountryCode']]
df_2018_2019 = df_2018_2019[['DateShort', 'Value', 'CountryCode']]
df_2019 = df_2019[['DateShort', 'Value', 'CountryCode']]
df_2020 = df_2020[['DateShort', 'Value', 'CountryCode']]
df_2021 = df_2021[['DateShort', 'Value', 'CountryCode']]
df_2022 = df_2022[['DateShort', 'Value', 'CountryCode']]
df_2023 = df_2023[['DateShort', 'Value', 'CountryCode']]
df_2024 = df_2024[['DateShort', 'Value', 'CountryCode']]
df_2025 = df_2025[['DateShort', 'Value', 'CountryCode']]
```

```python
# Vamos a ver si la casilla DateShort está en formato datetime

print(df_2015_2017['DateShort'].dtype)
print(df_2018_2019['DateShort'].dtype)
print(df_2019['DateShort'].dtype)
print(df_2020['DateShort'].dtype)
print(df_2021['DateShort'].dtype)
print(df_2022['DateShort'].dtype)
print(df_2023['DateShort'].dtype)
print(df_2024['DateShort'].dtype)
print(df_2025['DateShort'].dtype)

# Están ya en formato datetime, no tenemos que hacer nada más.
```

```python
# El archivo de 2018-2019 tiene un problema: los datos de 2019 están repetidos en el archivo de 2019. Vamos a eliminar esos datos del archivo de 2018-2019 (que además son incompletos)

df_2018_2019 = df_2018_2019[df_2018_2019['DateShort'] < '2019-01-01']

# El archivo de 2015 está casi incompleto.

df_2015_2017 = df_2015_2017[df_2015_2017['DateShort'] >= '2016-01-01']
```

```python
# Ahora vamos a hacer un único dataframe con todos los datos concatenados

df_todos = pd.concat([df_2015_2017, df_2018_2019, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024, df_2025], ignore_index=True)
```

```python
# Ahora nos interesa quedarnos solamente con los datos agrupados por mes y año. Por lo tanto, como tenemos varias filas por mes (una por cada hora del mes) vamos a hacer un groupby para agrupar por mes y año, y luego hacer una suma de los valores de cada mes.

df_todos['YearMonth'] = df_todos['DateShort'].dt.to_period('M')

df_todos.head()


```

```python
# Ahora vamos a agrupar por 'YearMonth' y 'CountryCode' y sumar los valores de 'Value'

df_mensual = df_todos.groupby(['YearMonth', 'CountryCode']).agg({'Value': 'sum'}).reset_index()
```

```python
# Ahora vamos a exportar el dataframe resultante a un archivo csv
df_mensual.to_csv(ruta_archivo + 'datos_electricidad_mensual.csv', index=False)
```

```python
# Ahora vemos la longitud del los datos para cada countrycode
print(df_mensual['CountryCode'].value_counts())
```
